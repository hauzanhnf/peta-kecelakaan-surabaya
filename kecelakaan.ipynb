```python
kskk = pd.read_csv("KSKK.csv")
kskk
```

```python
loc = kskk['Lokasi'].value_counts().reset_index()

loc.columns = ['Lokasi', 'Jumlah Kecelakaan']
loc['Kecamatan'] = loc['Lokasi'].str.replace('KEC. ', '', regex=False)
loc['Kecamatan'] = loc['Kecamatan'].str.replace(', KOTA SURABAYA', '', regex=False)
loc['Kecamatan'] = loc['Kecamatan'].str.title()

mapping_kecamatan = {
    'Karangpilang': 'Karang Pilang',
    'Pabean Cantikan': 'Pabean Cantian',
    'Sukomanunggal': 'Suko Manunggal',
    'Asem Rowo': 'Asemrowo',
    'Dukuhpakis': 'Dukuh Pakis'
}
loc['Kecamatan'] = loc['Kecamatan'].replace(mapping_kecamatan)
loc[['Kecamatan','Jumlah Kecelakaan']]
```

```python
kec_sby = loc['Kecamatan'].unique()
kec_sby
```

K-Means

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = loc[['Jumlah Kecelakaan']]
# Menentukan jumlah cluster optimal
inertia = []
K = range(1, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Plot Elbow Method
plt.figure(figsize=(8,5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Jumlah Cluster (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
X = loc[['Jumlah Kecelakaan']]
kmeans = KMeans(n_clusters=3, random_state=42)
loc['Cluster'] = kmeans.fit_predict(X)

# Tentukan zona berdasarkan rata-rata kecelakaan tiap cluster
cluster_order = loc.groupby('Cluster')['Jumlah Kecelakaan'].mean().sort_values(ascending=False).index

kategori_map = {}
for idx, cluster_label in enumerate(cluster_order):
    if idx == 0:
        kategori_map[cluster_label] = 'Zona Merah (Rawan)'
    elif idx == 1:
        kategori_map[cluster_label] = 'Zona Kuning (Waspada)'
    else:
        kategori_map[cluster_label] = 'Zona Hijau (Aman)'

loc['Kategori'] = loc['Cluster'].map(kategori_map)

# Merge cluster dan kategori ke GeoDataFrame df_sby
df_merge = df_sby.merge(loc[['Kecamatan', 'Jumlah Kecelakaan', 'Cluster', 'Kategori']],
                        left_on='ADM3_EN', right_on='Kecamatan', how='left')

# Hitung Longitude dan Latitude dari centroid polygon
df_merge['Longitude'] = df_merge.geometry.centroid.x
df_merge['Latitude'] = df_merge.geometry.centroid.y

# Susun urutan kolom sesuai permintaan
df_result = df_merge[['Kecamatan', 'Longitude', 'Latitude', 'Jumlah Kecelakaan', 'Cluster', 'Kategori']]

# Urutkan dari jumlah kecelakaan tertinggi ke terendah
df_result = df_result.sort_values(by='Jumlah Kecelakaan', ascending=False).reset_index(drop=True)
df_result
```

```python
temp_map = df_sby.set_index('ADM3_EN')

# Buat dataframe loc jadi punya index Kecamatan dan kolom yang lengkap
temp_kskk = loc.set_index('Kecamatan')[['Jumlah Kecelakaan', 'Cluster', 'Kategori']]

# Join data kecamatan ke peta
df_final = temp_map.join(temp_kskk, how='left')

# Hitung Longitude dan Latitude dari geometry (centroid tiap kecamatan)
df_final['Longitude'] = df_final.geometry.centroid.x
df_final['Latitude'] = df_final.geometry.centroid.y

# Reset index agar kolom 'ADM3_EN' kembali menjadi kolom biasa
df_final = df_final.reset_index()

# Drop kecamatan yang tidak punya data kecelakaan
df_final = df_final.dropna(subset=['Jumlah Kecelakaan'])
df_final.head()
```

```python
df_final.plot(column='Jumlah Kecelakaan', cmap='RdYlGn_r', legend=True)
```

```python
df_final.explore(column='Jumlah Kecelakaan', cmap='RdYlGn_r', legend=True, tooltip=['ADM3_EN', 'Longitude', 'Latitude', 'Jumlah Kecelakaan', 'Cluster', 'Kategori'])
```
